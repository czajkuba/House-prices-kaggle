Primary feature selection was done in a manual manner. Each feature was assigned a grade from scale 0-5. The logic used for this ranking was questioning each feature how strongly could if affect the price and whether if I were to buy a house, would I care about this feature?

```python
X_train1 = X_train[feats_man]
X_train1.GarageType.fillna(X_train.GarageType.mode()[0], inplace=True)
```

```python
# Need to assign a numerical value to all categorical features
for cat_feat in feats_cat:
    X_train1[str(cat_feat)+'_cat'] = pd.factorize(X_train1[cat_feat])[0]
X_train1.columns
```

```python
def run_RF(feats, x_train, y_train):
    X_train_predict = x_train[feats]
    X_train, X_test, Y_train, Y_test = train_test_split(X_train_predict, y_train, random_state=42)
    rf_regrr = RandomForestRegressor(n_estimators=100, random_state=42)
#     cv_k = KFold(n_splits=5, random_state=42)
    cv_score = cross_val_score(rf_regrr, X_train_predict, y_train, cv=5)
    print('CV 5 fold mean score: {} mean std: {}'.format(round(np.mean(cv_score),3), round(np.std(cv_score),3)))
#     rf_regrr.fit(X_train, y_train)
#     y_predict = rf_regrr.predict(X_test)
#     print(r2_score(y_test, y_predict))
```

```python
def run_RF_scaled(feats, x_train, y_train):
    X_train_predict = x_train[feats]
    x_train = StandardScaler().fit_transform(X_train_predict)
    y_train = StandardScaler().fit_transform(y_train.values.reshape(-1,1))
    rf_regrr = RandomForestRegressor(n_estimators=100, random_state=42)
#     cv_k = KFold(n_splits=5, random_state=42)
    cv_score = cross_val_score(rf_regrr, x_train, y_train, cv=5)
    print('CV 5 fold mean score w scaling: {} mean std: {}'.format(round(np.mean(cv_score),3), round(np.std(cv_score),3)))
```

```python
feats_man = X_train1.columns.values
feats_man2 = ['OverallQual', 'TotalBsmtSF', 'GarageArea', 'YearBuilt_cat', 'GrLivArea']
run_RF(feats_man2, X_train1, Y_train)
run_RF_scaled(feats_man2, X_train1, Y_train)
```
